```python
# Корреляция — статистическая мера, которая показывает степень и направление связи между двумя переменными.

# Коэффициент корреляции:
# Измеряет степень линейной связи между двумя переменными.
# Принимает значения: от -1 до 1.

# Ковариация — это мера, показывающая, как две переменные изменяются совместно. Она оценивает только направление линейной связи между переменными

# Детерминация - метрика, которая показывает, какая доля дисперсии зависимой переменной (y) объясняется моделью регрессии (объясняет данные) 
# Принимает значения: от 0 до 1.
# R = corr(x, y)^2

# Для корреляции Пирсона важны два ключевых условия: линейность и монотонность
```

```python
import random
import numpy as np
import matplotlib.pyplot as plt
from statistics import correlation, covariance

# р - настолько существенным будет случайный фактор
def random_shape(arr, p):
    y = np.zeros_like(arr)
    for i in range(arr.size):
        sign = 1 if random.choice([0, 1]) else -1
        y[i] = arr[i] + p * sign * random.random()
    return y

x = np.arange(-1, 1, 0.05)
y = random_shape(x, 1)
y1 = random_shape(x, 0.7)
y2 = random_shape(x, 0.1)

plt.figure(figsize=(20, 5))

plt.subplot(1, 3, 1)
cov = covariance(x, y)
corr = correlation(x, y)
print('Низкая корреляция')
print(f'Ковариация: {cov}')
print(f'Корреляция: {corr}', end='\n')
plt.scatter(x, y)
plt.title('Низкая корреляция')

plt.subplot(1, 3, 2)
cov = covariance(x, y1)
corr = correlation(x, y1)
print('Средняя корреляция')
print(f'Ковариация: {cov}')
print(f'Корреляция: {corr}', end='\n')
plt.scatter(x, y1)
plt.title('Средняя корреляция')

plt.subplot(1, 3, 3)
cov = covariance(x, y2)
corr = correlation(x, y2)
print('Высокая корреляция')
print(f'Ковариация: {cov}')
print(f'Корреляция: {corr}', end='\n')
plt.scatter(x, y2)
plt.title('Высокая корреляция')

plt.show()
```

```python
# Корреляция Пирсона измеряет линейную зависимость между переменными. Требует нормальности данных и линейности связи. Неустойчива к выбросам

# Корреляция Спирмена измеряет монотонную зависимость, не обязательно линейную. Не требует нормальности данных. Более устойчива к выбросам
```
