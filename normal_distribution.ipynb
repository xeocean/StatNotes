```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Нормальное распределение — описывает распределение случайных величин, которые склонны концентрироваться вокруг среднего значения, с постепенным уменьшением вероятности отклонений от этого среднего. 

mu = 0  # Среднее
sigma = 1  # Стандартное отклонение
size = 1000  # Размер выборки

data = np.random.normal(mu, sigma, size)
# data = np.random.randn(1000) sigma = 1 | mu = 0
plt.hist(data, bins=100)
plt.show()
```

```python
# Стандартизация - процесс преобразования случайной величины из произвольного нормального распределения с параметрами μ (среднее) и σ (стандартное отклонение) в стандартное нормальное распределение (среднее = 0, стандартное отклонение = 1)

# Cтандартизированное значение (Z) = Наблюдаемое значение - среднее значение распределения / стандартное отклонение

mu = 50      # Среднее значение
sigma = 10   # Стандартное отклонение
size = 1000  # Размер выборки

data = np.random.normal(mu, sigma, size)

z_data = (data - mu) / sigma

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.hist(data, bins=100, color='r')
plt.title('Histogram of data')
plt.subplot(1, 2, 2)
plt.hist(z_data, bins=100, color='b')
plt.title('Histogram of z_data')
plt.show()
```

```python
# Правило трех сигм для нормального распределения:

# 68% данных находятся в пределах одного стандартного отклонения от среднего (μ ± σ).
# 95% данных находятся в пределах двух стандартных отклонений от среднего (μ ± 2σ).
# 99.7% данных находятся в пределах трех стандартных отклонений от среднего (μ ± 3σ).

mu = 0  # Среднее
sigma = 1  # Стандартное отклонение
size = 1000  # Размер выборки

data = np.random.normal(mu, sigma, size)

# Вычисление границ для 1, 2 и 3 сигм
sigma_1 = [mu - sigma, mu + sigma]
sigma_2 = [mu - 2 * sigma, mu + 2 * sigma]
sigma_3 = [mu - 3 * sigma, mu + 3 * sigma]

plt.hist(data, bins=100)
plt.axvline(sigma_1[0], color='r')
plt.axvline(sigma_1[1], color='r')
plt.axvline(sigma_2[0], color='y')
plt.axvline(sigma_2[1], color='y')
plt.axvline(sigma_3[0], color='g')
plt.axvline(sigma_3[1], color='g')
plt.show()
```

```python
# Подсчет данных, которые лежат в пределах сигм

count_sigma_1 = round(len(data[(data > sigma_1[0]) & (data < sigma_1[1])]) / size * 100, 2)
count_sigma_2 = round(len(data[(data > sigma_2[0]) & (data < sigma_2[1])]) / size * 100, 2)
count_sigma_3 = round(len(data[(data > sigma_3[0]) & (data < sigma_3[1])]) / size * 100, 2)

print(count_sigma_1, count_sigma_2, count_sigma_3)
```
