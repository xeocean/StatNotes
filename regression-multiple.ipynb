```python
# Множественная линейная регрессия используется, когда есть несколько независимых переменных для предсказания одной зависимой переменной.

# Уравнение вида: y = b + a1 * x1 + a2 * x2 + ... + an * xn
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LinearRegression

df = pd.read_csv('data/states.csv')

X = df[['metro_res', 'white', 'hs_grad', 'female_house']]
y = df['poverty']

model = LinearRegression()
model.fit(X, y)

score = model.score(X, y)
y_pred = model.predict(X)

print(f'Пересечение с осью y: {model.intercept_}')
print(f'Коэффициенты наклона: {model.coef_}')
print(f'Коэффициент детерминации: {score}')

sns.pairplot(df, diag_kind='kde', kind='scatter')
```

```python
df = df.drop('state', axis=1)
```

```python
sns.heatmap(df.corr(), cmap='twilight', annot=True)
plt.title('Корреляция между переменными')
plt.show()
```

```python
# Adjusted R-squared (скорректированный коэффициент детерминации) — это модификация R², которая учитывает количество переменных в модели и размер выборки. Он используется, чтобы оценить, насколько хорошо модель объясняет данные, при этом учитывая количество использованных признаков.

r2 = model.score(X, y) # Коэффициент детерминации R²
n = X.shape[0]  # Количество наблюдений
p = X.shape[1]  # Количество признаков
r2_adjusted = 1 - (1 - r2) * (n - 1) / (n - p - 1)

print(f'R²: {r2}')
print(f'Adjusted R²: {r2_adjusted}')
```
