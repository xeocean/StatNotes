```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t
from scipy.stats import norm
```

```python
# t-распределение — это распределение вероятностей, которое используется, когда размер выборки мал, а стандартное отклонение неизвестно. Оно похоже на нормальное распределение, но имеет более "тяжелые хвосты". Это означает, что t-распределение более чувствительно к экстремальным значениям.

# Степени свободы (df) — это параметр, который определяет форму t-распределения. Чем больше степеней свободы, тем ближе t-распределение к нормальному (df = n - 1)
```

```python
x = np.arange(-5, 5, 0.01)
df = [1, 3, 5, 10, 20]

y = norm.pdf(x)
plt.plot(x, y)

for _ in df:
    plt.plot(x, t.pdf(x, df=_))
```

```python
# При увеличении степени свободы (например, при увеличении размера выборки) доверительный интервал для среднего значения становится ужже

mu = 89.9  # Среднее значение
sd = 11.3  # Стандартное отклонение
n = 20  # Размер выборки
p = 0.95  # Доверительная вероятность
alpha = 1 - p  # Уровень значимости
df = n - 1  # Степени свободы
SE = sd / n ** 0.5  # Стандартная ошибка

# Критическое значение t
interval =  t.ppf(1 - (alpha / 2), df)

print(f'{mu - interval * SE:.2f}, {mu + interval * SE:.2f}')
```

```python
# Данные
mu_m, mu_f = 45, 34
sd_m, sd_f = 9, 10
n = 100

# Расчет стандартной ошибки
se = (sd_m ** 2 / n + sd_f ** 2 / n) ** 0.5

# Расчет t-статистики
t_value = (mu_m - mu_f) / se

# Степени свободы
df = 2 * n - 2  # для независимых выборок

# Критическое значение t для двустороннего теста с уровнем значимости 0.05
alpha = 0.05
t_critical = t.ppf(1 - alpha / 2, df)

# Печать результатов
print(f"t-значение: {t_value:.2f}")
print(f"Критическое t-значение для уровня значимости {alpha}: {t_critical:.2f}")

# Сравнение t-значения с критическим
if abs(t_value) > t_critical:
    print("Гипотеза о равенстве средних отвергнута.")
else:
    print("Гипотеза о равенстве средних не отвергнута.")
```
